```python
from icecream import ic
from matplotlib import pyplot as plt
import numpy as np
```

```python

X = np.array([[51, 55], [14, 19], [0, 4]])

ic(X.flatten())
ic(X > 15)
ic(X[X>15])
ic(X[0])
ic(X[0,1])
```

```python
x = np.arange(0, 6, 0.1)
y = np.sin(x)

ic(x)
plt.plot(x, y)
```

```python
from cProfile import label


x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin')
plt.plot(x, y2, linestyle='--', label='cos') # cos 함수는 점선으로 그리기
plt.xlabel('x') # x축 이름
plt.ylabel('y') # y축 이름
plt.title('sin&cos') # 제목
plt.legend()
plt.show()
```
